**applying the different formatting lines**

```python
#format of the line-->doted (:)
#solid (-)
#---dashed lines
#dash dot line (-.)
plt.plot(x,y,"p:g")
plt.show()
```

```python
#change the size of the marker-->use parameter "marker size"-->ms
plt.plot(x,y,"p:g",markersize=10)
plt.show()
```

```python
#change the border of the marker to different colour
#use parameter called 'maker edge colour'-->mec
plt.plot(x,y,"p:g",mec='b')
plt.show()
```

```python
#change the marker colour to yellow --> use the parameter called 'marker fill colour' -->mfc
#use colour picker to give different colours
plt.plot(x,y,"p-.g",ms=10,mfc='y')
plt.show()
```

```python
#write a code to generate a plot with dash dot linesand hexagonal marker with marker face color is red and line colour is bluie
plt.plot(x,y,"p-.b",ms=10,mfc='r')
plt.show()
```

```python
x = [1,2,3,4,5,6,7,8,9]
y = [2,1,4,6,4,5,8,9,10]
plt.plot(x,y,'p')
plt.show()
```

```python
x = [1,2,3,4,5,6,7,8,9]
y = [2,1,4,6,4,5,8,9,10]
plt.plot(x,y,'p')
plt.xlabel('enter number')
plt.ylabel('marks in exam')
plt.title('marks vs entry number')
plt.grid()
plt.show()
#draw the grid only for x-axis -->plt.grid(x-axis='x')
plt.show()
```

```python
#how can we change line style and line width and colour of the grids
x = [1,2,3,4,5,6,7,8,9]
y = [2,1,4,6,4,5,8,9,10]
plt.plot(x,y,'p')
plt.xlabel('enter number')
plt.ylabel('marks in exam')
plt.title('marks vs entry number')
plt.grid(linestyle = '--',linewidth=1,colour='03e3fe')
#draw the grid only for x axis -->plt.grid(axis ='x)
plt.show()
```
