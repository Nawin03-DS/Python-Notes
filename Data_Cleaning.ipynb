```python
from ast import literal_eval
#if the variable == variable returns false then it means the variable is a nan values
jasmine = []
for i in data['belongs_to_collection']:
    if i == i:
        jasmine.append(ast.literal_eval(i)['name'])
    else:
        jasmine.append('No Collection')
```

```python
print(jasmine)
```

```python
data['collection_name'] = jasmine
```
